# SKLEARN

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df=pd.read_excel("Placement_Package_Dataset.xlsx")
df.head()
```

```python
x=df[["Aptitude","Coding","Communication","Projects"]]
print(x)
```

```python
y=df["Package_LPA"]
```

```python
x_train,X_test,y_train,y_test=train_test_split(
    x,y,train_size=0.25,random_state=1
)
print(x_train)
print(y_test)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print("R2 Score test=",round(model.score(X_test,y_test),2))
print("R2 Score train=",round(model.score(x_train,y_train),2))
```

```python
new_student=[[85,90,80,4]]
predicted_package=model.predict(new_student)
print("predicted_package=",round(predicted_package[0],2),)
```
