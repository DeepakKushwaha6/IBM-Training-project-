# 2. Word frequency counter

```python

sentence = "python is fun python is easy"
words = sentence.split()
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
print(freq)


```

# 3. Calculator with functions

```python
a=int(input("Enter the first number:-"))
b=int(input("Enter the second number:-"))
def add(a,b):
    return (f"{a} + {b} = {a+b}")
def sub(a,b):
    return (f"{a} + {b} = {a-b}")
def mult(a,b):
    return (f"{a} + {b} = {a*b}")
def divi(a,b):
    return (f"{a} + {b} = {a/b}")
print(add(a,b))
print(sub(a,b))
print(mult(a,b))
print(divi(a,b))

```

# 4. Marks analyser

```python

marks = [85, 72, 90, 40, 55]
minimum=min(marks)
maximum=max(marks)
average=sum(marks)/len(marks)

passed=len([m for m in marks if m >=50])
passed_percentage=(passed / len(marks))*100

grade=[]
for m in marks:
    if m>=80:
        grade.append("A")
    elif m >=60:
        grade.append("B")
    elif m >=50:
        grade.append("C")
    else:
        grade.append("F")

print("Min:", minimum)
print("Max:", maximum)
print("Avg:", average)
print("Pass %:", passed_percentage)
print("Grades:", grade)

```

# 5. Top-N students


```python

students = [
    {"name": "ram", "marks": 85},
    {"name": "rohan", "marks": 92},
    {"name": "rajveer", "marks": 78},
    {"name": "Delip", "marks": 95}
]
n = 2
top_students = sorted(
    students,
    key=lambda x: x["marks"],
    reverse=True
)[:n]
print(top_students)

```

# 1. Flatten a nested list
 ## convert [[1,2],[3,4],[5,6]] into [1,2,3,4,5,6]

```python

# 1. Flatten a nested list
 # convert [[1,2],[3,4],[5,6]] into [1,2,3,4,5,6]
nested = [[1, 2], [3, 4], [5, 6]]

flat = [item for sublist in nested for item in sublist]

print(flat)
```
