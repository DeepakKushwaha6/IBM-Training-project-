#### 1. Load a fresh copy of employees.csv. 
#### 2. Clean all missing values (Salary mean, Experience-0). 
#### 3. Add a Seniority column: 'Senior if Experience>= 5, else 'Junior 
#### 4. Show the average Salary for Seniors vs Juniors.

```python
df = pd.read_csv("employees.csv")
df.head()
```

```python
# Fill missing values
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Experience"] = df["Experience"].fillna(0)
```

```python
# Add Seniority column
seniority=[]
for x in df['Experience']:
    if x>=5:
        seniority.append('Senior')
    else:
        seniority.append('Junior')
df['seniority']=seniority
```

```python
# Show updated dataset
print(df)
```

```python
# Average salary by Seniority
print("\nAverage Salary by Seniority:")
print(df.groupby("seniority")["Salary"].mean())
```
