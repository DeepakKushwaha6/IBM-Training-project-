# PANDAS ASSIDNMENT

```python
import pandas as pd
import numpy as np

d = {
    "Name": ["Aman", "Priya", "Rohit", "Neha", "Karan", "Sara", "Vikram", "Meera"],
    "Department": ["Sales", "IT", "IT", "HR", "Sales", "IT", "HR", "Sales"],
    "Age": [28, 34, 25, 41, 30, 38, 29, 33],
    "Salary": [42000, 68000, 40000, np.nan, 51000, 72000, 45000, 58000],
    "Experience": [3, 8, 2, 15, 5, np.nan, 4, 7]
}

df = pd.DataFrame(d)

print(df)
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Experience"] = df["Experience"].fillna(0)

print(df)
```

```python
df["Bonus"] = df["Salary"] * 0.10

print(df)
```

# LEVEL 3 — CHALLENGE  

```python
print(df.groupby("Department")["Salary"].mean())
```

```python
print(df.groupby("Department")["Name"].count())
```

```python
# returns the index (department name) with the highest value.
print(df.groupby("Department")["Salary"].mean().idxmax())
```

```python
print("Highest Salary:", df["Salary"].max())
print("Lowest Salary:", df["Salary"].min())
print("Average Age:", df["Age"].mean())
```

```python
s = df.groupby("Department").agg({
    "Age": "mean",
    "Salary": "mean",
    "Name": "count"
})

print(s)
```

The HR department appears to be the most experienced, with the highest average experience among its employees. The IT department earns the highest average salary, indicating that technical roles may be paid more in this company. If I were hiring, I would consider both experience and skills rather than salary alone, as different departments have different responsibilities. However, this dataset is very small, so it may not represent real-world trends accurately; an AI model trained on such limited data could produce biased or unreliable conclusions.
